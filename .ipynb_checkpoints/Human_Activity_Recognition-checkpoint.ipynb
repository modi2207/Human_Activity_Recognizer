# 1) Plot the waveform for data from each activity class. Are you able to see any difference/similarities between the activities? You can plot a subplot having 6 colunms to show differences/similarities between the activities. Do you think the model will be able to classify the activities based on the data? [1 marks]



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Corrected base path to the local C drive
base_path = "Plot_Data"
save_dir = "Figures"  # Added save directory path
subjects = ['laying', 'sitting', 'standing', 'walking', 'down', 'up']

# Create the save directory if it doesn't exist
os.makedirs(save_dir, exist_ok=True)

# Create the subplot figure
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
axes = axes.flatten()

# Plot waveforms for each subject
for idx, subject in enumerate(subjects):
    file_path = os.path.join(base_path, f"{subject}.csv")
    data = pd.read_csv(file_path, header=None)

    ax = axes[idx]
    x = np.arange(1, 500, 2)
    y_x = [float(data.iloc[i, 0]) for i in x]
    y_y = [float(data.iloc[i, 1]) for i in x]
    y_z = [float(data.iloc[i, 2]) for i in x]

    ax.plot(x, y_x, color='red', label='X-axis')
    ax.plot(x, y_y, color='blue', label='Y-axis')
    ax.plot(x, y_z, color='green', label='Z-axis')

    ax.set_title(f"Waveform for {subject}")
    ax.set_xlabel("Time")
    ax.set_ylabel("Acceleration")
    ax.legend()

# Adjust layout and save the plot
plt.tight_layout()
save_path = os.path.join(save_dir, "Q1_all_activities_plot.png")
plt.savefig(save_path)

# Optional: Display the plot
plt.show()
```

 Amplitude helps distinguish activities based on the intensity of their movements. The higher the amplitude peaks, the greater the force exerted, suggesting activities like walking upstairs or downstairs. Conversely, lower amplitudes indicate static movements like sitting standing.

By analyzing the amplitude variations in each axis, we can gain valuable insights into the dynamics of different activities and potentially use this information for activity recognition tasks with machine learning models.

# 2)Do you think we need a machine learning model to differentiate between static activities (laying, sitting, standing) and dynamic activities(walking, walking_downstairs, walking_upstairs)? Look at the linear acceleration $ (acc_x^2+acc_y^2+acc_z^2) $ for each activity and justify your answer. [1 mark]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Corrected base path to the local C drive
base_path = "Plot_Data"
save_dir = "Figures"  # Added save directory path
subjects = ['laying', 'sitting', 'standing', 'walking', 'down', 'up']
# Create the save directory if it doesn't exist
os.makedirs(save_dir, exist_ok=True)

# Create the subplot figure
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
axes = axes.flatten()

# Plot waveforms for each subject
for idx, subject in enumerate(subjects):
    file_path = os.path.join(base_path, f"{subject}.csv")
    data = pd.read_csv(file_path, header=None)

    ax = axes[idx]
    x = np.arange(1, 500, 2)
    y_x = [float(data.iloc[i, 0])**2 for i in x]  # Squared values for X-axis
    y_y = [float(data.iloc[i, 1])**2 for i in x]  # Squared values for Y-axis
    y_z = [float(data.iloc[i, 2])**2 for i in x]  # Squared values for Z-axis

    y = [y_x[i] + y_y[i] + y_z[i] for i in range(len(y_x))]  # Sum of squared values

    ax.plot(x, y, color='red', label='Sum of Squared Acceleration')

    ax.set_title(f"Waveform for {subject}")
    ax.set_xlabel("Time")
    ax.set_ylabel("Sum of Squared Acceleration")
    ax.legend()

# Adjust layout and save the plot
plt.tight_layout()
save_path = os.path.join(save_dir, "Q2_all_activities_plot.png")
plt.savefig(save_path)

# Optional: Display the plot
plt.show()
```

We think we do not need any machine learning model to distinguish between between static activities (laying, sitting, standing) and dynamic activities(walking, walking_downstairs, walking_upstairs). Using linear acceleration , we can clearly distinguish between static and dynamic activities. For static activities , we found amplitude to be of lower magnitude, whereas due to involvement of larger force for dynamic activities, we clearly see amplitude of higher magnitude. 

# 3) Train Decision Tree using trainset and report Accuracy and confusion matrix using testset.

```python
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
#
#                                   ES335- Machine Learning- Assignment 1
#
# This file is used to create the dataset for the mini-project. The dataset is created by reading the data from
# the Combined folder. The data is then split into training, testing, and validation sets. This split is supposed
# to be used for all the modeling purposes.
#
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Library imports
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
import os
import pickle

# Constants
time = 10
offset = 100
folders = ["LAYING","SITTING","STANDING","WALKING","WALKING_DOWNSTAIRS","WALKING_UPSTAIRS"]
classes = {"WALKING":1,"WALKING_UPSTAIRS":2,"WALKING_DOWNSTAIRS":3,"SITTING":4,"STANDING":5,"LAYING":6}

combined_dir = os.path.join("Combined")

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Train Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

X_train=[]
y_train=[]
dataset_dir = os.path.join(combined_dir,"Train")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))

    for file in files:

        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50]
        X_train.append(df.values)
        y_train.append(classes[folder])

X_train = np.array(X_train)
y_train = np.array(y_train)


#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Test Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

X_test=[]
y_test=[]
dataset_dir = os.path.join(combined_dir,"Test")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))
    for file in files:

        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50]
        X_test.append(df.values)
        y_test.append(classes[folder])

X_test = np.array(X_test)
y_test = np.array(y_test)

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Final Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# USE THE BELOW GIVEN DATA FOR TRAINING, TESTING, AND VALIDATION purposes

# concatenate the training and testing data
X = np.concatenate((X_train,X_test))
y = np.concatenate((y_train,y_test))

# split the data into training,testing, and validation sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=4,stratify=y)
X_test,X_val,y_test,y_val = train_test_split(X_test,y_test,test_size=0.5,random_state=4,stratify=y_test)

print("Training data shape: ",X_train.shape)
print("Testing data shape: ",X_test.shape)
print("Validation data shape: ",X_val.shape)

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=


```

# 4) Train Decision Tree with varrying depths (2-8) using trainset and report accuracy and confusion matrix using Test set. Does the accuracy changes when the depth is increased? Plot the accuracies and reason why such a result has been obtained

```python
accuracy=[]
    
```

```python
def find(X_train,y_train,X_test,y_test,depth):
    model = DecisionTreeClassifier(max_depth=depth,criterion='entropy')
    model.fit(X_train, y_train)
    y_predict=model.predict(X_test)
    df=pd.DataFrame(np.zeros((6,6),int),columns=['WALKING','WALKING_UPSTAIRS','WALKING_DOWNSTAIRS','SITTING','STANDING','LAYING'],index=['WALKING','WALKING_UPSTAIRS','WALKING_DOWNSTAIRS','SITTING','STANDING','LAYING'])
    
    for i in range(0,len(y_predict)):
        df.iloc[y_predict[i]-1,y_test[i]-1]=df.iloc[y_predict[i]-1,y_test[i]-1]+1
    df['PRECISION'] = np.zeros(len(df))
    for i in range(len(df)):
        sum=df.iloc[i,:].sum()
        df.iloc[i,6]=df.iloc[i,i]/sum
        
    df.loc['RECALL']=np.zeros(7,int)
    
    for i in range(len(df.columns)):
        sum=df.iloc[:,i].sum()
        df.iloc[6,i]=df.iloc[i,i]/sum
    
    total_sum=0
    sum=0
    for i in range(len(df.columns)):
        total_sum=total_sum+df.iloc[:,i].sum()
    for i in range(len(df.columns)):
        sum=sum+df.iloc[i,i]
    accur=sum/total_sum
    
    print("Depth : ",depth)
    
    print(df)
    print("Accuracy: ",accur)
    accuracy.append(accur)
    print("\n\n\n\n\n")
    
```

```python

for depth in range(2,9):
    find(X_train,y_train,X_test,y_test,depth)
```

```python
import matplotlib.pyplot as plt

X=np.arange(2,9,1)

plt.plot(X,accuracy,marker="o")
plt.xlabel("DEPTH")
plt.ylabel("ACCURACY")
plt.show()
```

 As we increase the depth of decision tree,  it allows the model to learn more complex relationship with data. This leds the model to train well with training data. However after certain depth , model had overfitting issue. In overfitting model become too specific to training data and performs poorly on test data. Hence after depth 5 , accuracy decreases due to overfitting.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tsfel import time_series_features_extractor
from sklearn.decomposition import PCA
import os

data_directory = './Combined/Train/'
# Initialize empty lists to store features and labels
X_list = []
y_list = []


# Iterate through each activity folder
for activity_folder in os.listdir(data_directory):
    activity_path = os.path.join(data_directory, activity_folder)

    # Iterate through files in the activity folder
    for filename in os.listdir(activity_path):
        file_path = os.path.join(activity_path, filename)

        # Read the data from the file
        data = pd.read_csv(file_path)
        
        
        # Calculate Total Acceleration (acc_x^2 + acc_y^2 + acc_z^2)
        data['total_acc'] = data['accx']**2 + data['accy']**2 + data['accz']**2
        acceleration_data = data[['accx', 'accy', 'accz']].values
        
      
      
        # Append the Total Acceleration and activity label to the lists
        X_list.append(data[['accx', 'accy', 'accz']])
        y_list.append(activity_folder)

# Concatenate the lists to create the final dataset
X = pd.concat(X_list, axis=0, ignore_index=True)
y = pd.Series(y_list, name='activity')

# Apply PCA to compress the acceleration time series into two components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)


# Create a DataFrame with the PCA components and labels
pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Activity'] = y

# Plot scatter plot to visualize different classes of activities
plt.figure(figsize=(10, 8))
activities = y.unique()

for activity in activities:
    subset = pca_df[pca_df['Activity'] == activity]
    plt.scatter(subset['Principal Component 1'], subset['Principal Component 2'], label=activity)
    
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Total Acceleration for Different Activities')
plt.legend()
plt.show()

```
